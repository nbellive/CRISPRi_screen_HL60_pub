## Experiment - 20200319


This is my first attempt at a transwell assay,  where I'm trying to look at the number of differentiated HL60 cells that are able to migrate through the transwell membrane (toward a fMLP gradient). 

Note that I used a paper from the Lim lab (UCSF?), Park et al, PNAS, as a guide - with the expectation that 1 hr should be a good time point to see good migration. They performed transwell assays using 24 well, 3 $\mu m$ pores, with 100 nM fMLP in the bottom well and 1 hour migration time.  

Experimental setup:
* cells from 6x5ml flasks were spun down and resuspended into 6 ml RPMI (no FBS!). Cell count: 13*10^5 cells/ml.
* ~5 x 10^5 cells per incert, in 450 uL
* Time points: 30 min,  60 min, and 4 hr. 
 * duplicates, 1.5 ml in bottom well, 100 nM fMLP (note that I used a 1 mM stock I got from Rikki, prepared in 2016, in DMSO stored at ~-20C.
 * duplicates, 1.5 ml in bottom well, media only.
* Duplicates per condition. 

I first added media to bottom well, and then cells to top.

I'll look at the .FCS files in a moment, but for now, I also manualy gated the cells (based on running cells from the initial cell stock through the cytometer - which allowed me to distinguish 'dead' cells). Lets just plot these numbers. 

* Sample set t_0: 50 uL cells diluted to 1000 uL, analysis of 400 uL.
* Sample set t_30 min, 60 min: 1.2 mL counted.
* Sample set t_4 hr: 600. uL counted.
